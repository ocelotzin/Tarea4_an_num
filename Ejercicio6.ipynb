# Ejercicio 6

Vamos a calcular el punto fijo en el intervalo $[0, 2\pi]$ con una tolerancia de $10^{-2}$. A continuación el programa y su respectiva ejecución:

```python
import numpy as np
import matplotlib.pyplot as plt

# La función sobre la que vamos a trabajar:
def g(x):
    return np.pi + 0.5 * np.sin(x / 2)

# Valores necesarios
tol = 1e-2 # Tolerancia
maxVeces = 100 # Máximo de iteraciones
x0 = 2.0  # Valor inicial
valores_iterados = [x0]

# Iteración de Punto Fijo
for i in range(maxVeces):
    x1 = g(x0) # Hacemos el despeje
    valores_iterados.append(x1)
    if abs(x1 - x0) < tol: # Rompemos hasta encontrar la tolerancia deseada
        print(f"El punto fijo es: x ≈ {x1:.5f} para {i+1} iteraciones")
        break
    x0 = x1
else:
    print("No se encontró punto fijo con la tolerancia dada")

# Gráfica de la función g(x) y la línea y = x
x_vals = np.linspace(0, 2 * np.pi, 400)
y_vals = g(x_vals)

plt.plot(x_vals, y_vals, label='g(x) = π + 0.5·sin(x/2)', color='blue')
plt.plot(x_vals, x_vals, label='y = x', color='red', linestyle='--')
plt.scatter(valores_iterados, [g(x) for x in valores_iterados], color='green', zorder=5)
plt.title('Ejercicio 6')
plt.xlabel('x')
plt.ylabel('g(x)')
plt.legend()
plt.grid(True)
plt.show()
```
