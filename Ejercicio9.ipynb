```python
from sympy import symbols, sympify, diff, lambdify

# definimos x para que python lo pueda interpretar como una varible de manera global ya que sin esto el codigo falla
x = symbols('x')

def biseccion(a, b, f_expr, tol=1e-8, max_iter=1000):

    # Convertimos la función simbólica a una función numérica evaluable ademas anteriormente ya definimos x ya que por si sola esta da error
    f = lambdify(x, f_expr, "math")

    # Verificamos que haya un cambio de signo en los extremos del intervalo
    if f(a) * f(b) >= 0:
        return "Error: La función no cambia de signo en el intervalo dado."

    # Lista para guardar los puntos medios
    m_list = []

    # Iteramos hasta alcanzar la tolerancia o el número máximo de iteraciones
    for _ in range(max_iter):
        # Calculamos el punto medio del intervalo actual
        m = (a + b) / 2
        m_list.append(m)

        # Verificamos si la longitud del intervalo ya es menor que la tolerancia
        if abs(b - a) < tol:
            return m  # Raíz aproximada encontrada

        # Decidimos en qué subintervalo continuar según el cambio de signo
        if f(a) * f(m) < 0:
            b = m
        else:
            a = m

    # Si se llega aquí, no se alcanzó la tolerancia en el número máximo de iteraciones
    return "No se alcanzó la tolerancia deseada en el número máximo de iteraciones."
```

```python
def secante(x0, x1, f_expr, tol=1e-8, max_iter=1000):

    # Convertimos la función simbólica a una función evaluable numéricamente (Igual que en el anterior)
    f = lambdify(x, f_expr, "math")

    # Iteramos hasta alcanzar la tolerancia o el número máximo de iteraciones
    for _ in range(max_iter):
        f_x0 = f(x0)
        f_x1 = f(x1)

        # Evitamos división por cero o valores demasiado cercanos en el denominador
        if abs(f_x1 - f_x0) < 1e-14:
            return "Error: División por cero o valores muy cercanos en denominador."

        # Fórmula del método de la secante
        x2 = x1 - f_x1 * (x1 - x0) / (f_x1 - f_x0)

        # Condición de paro basada en la diferencia entre iteraciones
        if abs(x2 - x1) < tol:
            return x2  # Raíz aproximada encontrada

        # Actualizamos los valores para la siguiente iteración
        x0, x1 = x1, x2

    # Si no se alcanza la tolerancia en el número máximo de iteraciones
    return "No se alcanzó la tolerancia deseada en el número máximo de iteraciones."
```

```python
def newton(x0, f_expr, df_expr, tol=1e-8, max_iter=1000):

    f = lambdify(x, f_expr, "math")
    df = lambdify(x, df_expr, "math")

    # Iteramos hasta alcanzar la tolerancia o el número máximo de iteraciones
    for _ in range(max_iter):
        f_x0 = f(x0)      # Evaluamos la función en el punto actual
        df_x0 = df(x0)    # Evaluamos la derivada en el punto actual

        # Verificamos si la derivada es demasiado pequeña para evitar división por cero
        if abs(df_x0) < 1e-14:
            return "Error: Derivada cercana a cero. Método no aplicable."

        # Aplicamos la fórmula de Newton-Raphson
        x1 = x0 - f_x0 / df_x0

        # Verificamos si la diferencia entre iteraciones es menor que la tolerancia
        if abs(x1 - x0) < tol:
            return x1  # Se ha encontrado una raíz aproximada

        # Actualizamos el punto para la siguiente iteración
        x0 = x1

    # Si no se alcanza la tolerancia en el número máximo de iteraciones
    return "No se alcanzó la tolerancia deseada en el número máximo de iteraciones."
```
